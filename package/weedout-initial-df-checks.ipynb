```python
#Function to check if data loaded from a CSV file and data frame a CSV type

import pandas as pd
import os

def is_csv_file(filename):
    return filename.lower().endswith('.csv') #Checking if extension is.csv

def check_and_read_csv(file):
    try:
        df = pd.read_csv(file)
        return df, True #Return Dataframe and flag indicating it's a CSV
    except pd.errors.ParseError:
        return None, False #Return None and False flag if parsing fails
```

```python
#Function to check multiple same feature names

import pandas as pd

def find_duplicate_features(df):
    columns = df.columns.tolist() #Get a list of all column names
    
    duplicate_columns = {col for col in columns if columns.count(col) >1} #Find duplicates using a set of comphrehension
    return list(duplicate_columns)
```
